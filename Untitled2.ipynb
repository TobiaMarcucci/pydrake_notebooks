```python
import numpy as np
import matplotlib.pyplot as plt
from pydrake.all import MathematicalProgram, Polynomial, Variables, SolutionResult
```

```python
import numpy as np
from pympc.geometry.polyhedron import Polyhedron
from monomial import moment
```

```python
d_h = 4
d_f = 4
x_max = np.array([2.]*2)
x_min = - x_max
```

```python
nx = 2
ny = 1
n_cons = 20
#A = np.random.randn(n_cons, nx + ny)
#b = np.random.rand(n_cons, 1)
#np.save('A_random', A)
#np.save('b_random', b)
A = np.load('A_random.npy')
b = np.load('b_random.npy')
P = Polyhedron(A, b)
proj = P.project_to([0,1])
```

```python
prog = MathematicalProgram()
x = prog.NewIndeterminates(nx, 'x')
y = prog.NewIndeterminates(ny, 'y')
xy = np.concatenate([x,y])
```

```python
g = [-A[i,:].dot(xy) + b[i] for i in range(n_cons)]
```

d = prog.NewIndeterminates(n_cons, 'd')
h_dist =  [prog.NewSosPolynomial(Variables([d[i]]), d_h)[0] for i in range(n_cons)]
h = []
for i, hi in enumerate(h_dist):
    hi = hi.ToExpression()
    hi = hi.Substitute(d[i], g[i])
    h.append(Polynomial(hi))

```python
f = prog.NewFreePolynomial(Variables(x), d_f)
h =  [prog.NewSosPolynomial(Variables(xy), d_h)[0] for i in range(n_cons)]
```

```python
phi = 0.
for i in range(len(g)):
    phi += h[i].ToExpression()*g[i]
prog.AddSosConstraint(f.ToExpression() - phi)
```

```python
for c in f.decision_variables():
    prog.AddLinearConstraint(c <= 1.)
    prog.AddLinearConstraint(c >= -1.)
```

```python
obj = 0.
for mon, coef in f.monomial_to_coefficient_map().items():
    powers = mon.get_powers()
    order = [0]*nx
    for i, xi in enumerate(x):
        for xdict, power in powers.items():
            if xi.EqualTo(xdict):
                order[i] = power
    obj += moment(order, x_min, x_max) * coef
prog.AddLinearCost(obj)
```

```python
result = prog.Solve()
assert(result == SolutionResult.kSolutionFound)
```

```python
f_opt = prog.SubstituteSolution(f)
def f_eval(x_bar):
    return f_opt.Evaluate({xi: x_bar[i] for i, xi in enumerate(x)})
```

```python
fig = plt.figure()

n = 100
X = np.linspace(x_min[0], x_max[0], n)
Y = np.linspace(x_min[1], x_max[1], n)
X, Y = np.meshgrid(X, Y)

F = np.array([f_eval([p, q]) for p, q in zip(np.ravel(X), np.ravel(Y))]).reshape(n, n)
CS = plt.contour(X, Y, F, colors='r')
plt.clabel(CS)
plt.plot([0.], [.0], color='r', label=r'$f(x) = 0$')

proj.plot(label=r'$\pi_x(K)$')

plt.xlabel(r'$x$')
plt.ylabel(r'$y$, $f$')
plt.legend(loc=8)
plt.show()
```
