```python
import numpy as np
import matplotlib.pyplot as plt
from pydrake.all import MathematicalProgram, Variable, Variables, SolutionResult
```

```python
from monomial import moment
```

```python
d_h = 2
d_f = 4
x_max = 1.2
x_min = - x_max
```

```python
prog = MathematicalProgram()
x = prog.NewIndeterminates(1, 'x')[0]
y = prog.NewIndeterminates(1, 'y')[0]
```

```python
#g = [x + 1, 1 - x, y + 1, 1 - y]
g = [x**2-x**4-(y-.5)**2-.1, y-x**2]
#g = [(x+y)**2-(x+y)**4-(x-y)**2-.1, (x-y)-(x+y)**2+.5]
```

```python
f = prog.NewFreePolynomial(Variables([x]), d_f)
h =  [prog.NewSosPolynomial(Variables([x, y]), d_h)[0] for i in range(len(g))]
```

```python
phi = 0.
for i in range(len(g)):
    phi += h[i].ToExpression()*g[i]
prog.AddSosConstraint(f.ToExpression() - phi)
```

```python
for c in f.decision_variables():
    prog.AddLinearConstraint(c <= 1.)
    prog.AddLinearConstraint(c >= -1.)
```

```python
obj = 0.
for m, c in f.monomial_to_coefficient_map().items():
    assert(len(m.GetVariables()) <= 1)
    power = m.get_powers().values()
    if len(power) == 0:
        power = [0]
    obj += moment([power[0]], [x_min], [x_max]) * c
prog.AddLinearCost(obj)
```

```python
result = prog.Solve()
assert(result == SolutionResult.kSolutionFound)
```

```python
h_opt = [prog.SubstituteSolution(h_i) for h_i in h]
phi_opt = prog.SubstituteSolution(phi)
f_opt = prog.SubstituteSolution(f)
def phi_eval(p, q):
    return phi_opt.Evaluate({x: p, y: q})
def f_eval(p):
    return f_opt.Evaluate({x: p})
```

```python
def g_min(p, q):
    g_eval = [g_i.Evaluate({x: p, y: q}) for g_i in g]
    return min(g_eval)
```

```python
fig = plt.figure()

n = 100
X = np.linspace(x_min, x_max, n)
Y = np.linspace(x_min, x_max, n)
X, Y = np.meshgrid(X, Y)

PHI = np.array([phi_eval(p, q) for p, q in zip(np.ravel(X), np.ravel(Y))]).reshape(n, n)
G = np.array([g_min(p, q) for p, q in zip(np.ravel(X), np.ravel(Y))]).reshape(n, n)
CS = plt.contour(X, Y, PHI, colors='b')
plt.clabel(CS)
plt.contour(X, Y, G, colors='r', levels=[0.])

plt.plot([0.], [.0], color='r', label=r'$K$')
plt.plot([0.], [.0], color='b', label=r'$\sum_{i=1}^{n_g} h_i(x,y) g_i(x,y)$')

X = np.linspace(x_min, x_max, n)
F = [f_eval(x_i) for x_i in X]
plt.grid(True)
plt.plot(X, F, color='g', label=r'$f(x)$')

plt.xlim(x_min, x_max)
plt.ylim(x_min, x_max)

plt.xlabel(r'$x$')
plt.ylabel(r'$y$, $f$')
plt.legend(loc=8)
plt.show()
```
